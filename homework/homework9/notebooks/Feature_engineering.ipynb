## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, PolynomialFeatures

```

## GENERATING SYNTHETIC DATASET

```python
np.random.seed(42)
n = 200

df = pd.DataFrame({
    'customer_id': range(1, n+1),
    'income': np.random.normal(50000, 10000, n).astype(int),
    'credit_score': np.random.normal(720, 50, n).astype(int),
    'monthly_spend': np.random.normal(2000, 600, n).astype(int),
    'default_flag': np.random.choice([0,1], n, p=[0.8,0.2])
})
df.head()
```

```python
df['default_flag'].value_counts()
```

## BASIC TRANSFORMATION

```python
# Percent change in monthly spending
df['spend_pct_change'] = df['monthly_spend'].pct_change().fillna(0)

# Rolling mean of spending (3-month window)
df['rolling_spend_mean'] = df['monthly_spend'].rolling(3).mean().fillna(df['monthly_spend'])

# Binary flag for high spenders
df['high_spender_flag'] = (df['monthly_spend'] > 2500).astype(int)
df.head()
```

## INTERACTION, SQUARED, POLYNOMIAL FEATURES

```python
df['income'] = pd.to_numeric(df['income'], errors='coerce')
df['credit_score'] = pd.to_numeric(df['credit_score'], errors='coerce')
# Interaction feature
df['income_x_credit'] = df['income'] * df['credit_score']

# Squared feature
df['income_squared'] = df['income'] ** 2

# Polynomial features using sklearn
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df[['income','credit_score']])
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(['income','credit_score']))
df = pd.concat([df, poly_df], axis=1)
df.head()
```

```python
corr = df[['customer_id','income','credit_score','monthly_spend','default_flag']].corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
corr
```

## ADDITIONAL COMMENTS
i) Always ensure columns like income or credit_score are truly numbers (not strings with commas/spaces).
ii) Interaction features- Capture combined effects between variables.
iii)Polynomial features- Allow models to capture non-linear relationships between predictors and target.
